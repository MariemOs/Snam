```python
for zipcode in zipcodes:
    response = urllib2.urlopen('http://www.city-data.com/zips/'+str(zipcode)+'.html')
    html = response.read()

    m = re.search('<b>(Males):</b>.(\d*\,*\d*)', html)
    print(zipcode,m.group(1),(m.group(2).replace(',','')))
    
    m = re.search('<b>(Females):</b>.(\d*\,*\d*)', html)
    print(zipcode,m.group(1),(m.group(2).replace(',','')))
```
