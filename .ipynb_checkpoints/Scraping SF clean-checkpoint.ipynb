```python
import urllib2
import re
import pandas as pd
import csv
```

## Selection des zip codes de San Francisco

```python
zipcodes=(94102, 94103, 94104, 94105, 94107, 94108, 94109, 94110, 94111, 94112, 94114, 94115, 94116, 94117, 94118, 94121, 94122, 94123, 94124, 94127, 94129, 94130, 94131, 94132, 94133, 94134)
```

## Exctraction des données

```python
df2=pd.DataFrame([[]])
for zipcode in zipcodes:
    response = urllib2.urlopen('http://www.city-data.com/zips/'+str(zipcode)+'.html')
    html = response.read()
    
    m = re.search('<b>(Estimated zip code population in 2013):</b>.(.*)<br>', html)
    v1=(zipcode,m.group(1),int(m.group(2).replace(',','')))
    
    m = re.search('<b>(Zip code population in 2010):</b>.(.*)<br>', html)
    v2=(zipcode,m.group(1),int(m.group(2).replace(',','')))
    
    m = re.search('<b>(Zip code population in 2000):</b>.(.*)<br>', html)
    v3=(zipcode,m.group(1),int(m.group(2).replace(',','')))
    
    m = re.search('<b>(Houses and condos):</b>.(.*)<br>', html)
    v4=(zipcode,m.group(1),int(m.group(2).replace(',','')))    
    
    m = re.search('<b>(Renter-occupied apartments):</b>.(.*)<br>', html)
    v5=(zipcode,m.group(1),int(m.group(2).replace(',',''))) 
    
    m = re.search('<b>(Mar. 2016 cost of living index in zip code).*:</b>.(\d+.+?)<', html)
    v6=(zipcode,m.group(1),float(m.group(2).replace(',','')))
    
    m = re.search('<b>(Land area):</b>.(.*)<b>', html)
    v7=(zipcode,m.group(1),float(m.group(2).replace(',','')))   
    
    m = re.search('<b>(Water area):</b>.(.*)<b>', html)
    v8=(zipcode,m.group(1),float(m.group(2).replace(',','')))  
    
    m = re.search('<b>(Population density):</b>.(.*)<b>', html)
    v9=(zipcode,m.group(1)+' (people per square mile)',float(m.group(2).replace(',',''))) 
      
    m = re.search('<b>(Males):</b>.(\d*\,*\d*)', html)
    v12=(zipcode,m.group(1),int(m.group(2).replace(',','')))
    
    m = re.search('<b>(Females):</b>.(\d*\,*\d*)', html)
    v13=(zipcode,m.group(1),int(m.group(2).replace(',','')))
    
    m = re.search('>(\d*\,*\d*)</span>.*(White population)', html)
    v14=(zipcode,m.group(2),float(m.group(1).replace(',','')))
    
    m = re.search('>(\d*\,*\d*)</span>.*(Black population)', html)
    v15=(zipcode,m.group(2),float(m.group(1).replace(',','')))
    
    m = re.search('>(\d*\,*\d*)</span>.*(American Indian population)', html)
    v16=(zipcode,m.group(2),float(m.group(1).replace(',','')))
    
    m = re.search('>(\d*\,*\d*)</span>.*(Asian population)', html)
    v17=(zipcode,m.group(2),float(m.group(1).replace(',','')))
    
    m = re.search('>(\d*\,*\d*)</span>.*(Native Hawaiian and Other Pacific Islander population)', html)
    v18=(zipcode,m.group(2),float(m.group(1).replace(',','')))
    
    m = re.search('>(\d*\,*\d*)</span>.*(Some other race population)', html)
    v19=(zipcode,m.group(2),float(m.group(1).replace(',','')))
    
    m = re.search('>(\d*\,*\d*)</span>.*(Two or more races population)', html)
    v20=(zipcode,m.group(2),float(m.group(1).replace(',','')))
    
    m = re.search('>(\d*\,*\d*)</span>.*(Hispanic or Latino population)', html)
    v21=(zipcode,m.group(2),float(m.group(1).replace(',','')))
    
    m = re.search('<b>(Median resident age):\S*\s\S*\s\S*\s\S*\s\S{28,32}(\d+\.*\d*)', html)
    v22=(zipcode,m.group(1),float(m.group(2).replace(',','')))
    
    m = re.search('<b>(Average household size):\S*\s\S*\s\S*\s\S*\s\S{28,32}(\d+\.*\d*)', html)
    v23=(zipcode,m.group(1),float(m.group(2).replace(',','')))
    
    m = re.search('<b>(Average Adjusted Gross Income \(AGI\) in 2012):</b>.\$(\d*\,*\d*)', html)
    v24=(zipcode,m.group(1),int(m.group(2).replace(',','')))
    
    m = re.search('<b>(Salary/wage):</b>.\$(\d*\,*\d*)', html)   
    #attention : deux champs idéntiques pour 2012 et 2004, algo prend le premier champs, donc 2012
    v25=(zipcode,m.group(1),int(m.group(2).replace(',','')))
    
    m = re.search('<b>(Residents with income below 50% of the poverty level in 2013):\S*\s\S*\s\S*\s\S*\s\S*\s\S*\s\S*\s\S*\s\S{28,32}(\d+\.*\d*)', html)
    v26=(zipcode,m.group(1),float(m.group(2).replace(',','')))
    
    df_row=pd.DataFrame([{'ZipCode':str(int(zipcode)),v1[1]:v1[2],v2[1]:v2[2],v3[1]:v3[2],v4[1]:v4[2],v5[1]:v5[2],v6[1]:v6[2],v7[1]:v7[2],v8[1]:v8[2],v9[1]:v9[2],v12[1]:v12[2],v13[1]:v13[2],v14[1]:v14[2],v15[1]:v15[2],v16[1]:v16[2],v17[1]:v17[2],v18[1]:v18[2],v19[1]:v19[2],v20[1]:v20[2],v21[1]:v21[2],v22[1]:v22[2],v23[1]:v23[2],v24[1]:v24[2],v25[1]:v25[2],v26[1]:v26[2]}])
    df2 = df2.append(df_row)
    
```
